### 1. Proceso de Limpeza de Datos.

** Deteccin y correccin de datos duplicados.**

```python
#Contar registro Duplicados.
df_vintrafamiliar.duplicated().sum()
```

```python
#Deteccin y correccin de datos duplicados.
df_vintrafamiliar.drop_duplicates(inplace=True)   #Eliminar la consulta de los duplicados que se arroj en la linia de codigo [37].
df_vintrafamiliar
```

** Eliminacin de valores nulos.**

```python
#Eliminacin de valores nulos
df_vintrafamiliar.dropna(inplace=True)
df_vintrafamiliar
```

```python
# Eliminar filas con muchos valores nulos
df_vintrafamiliar.dropna(thresh=2, inplace=True)
df_vintrafamiliar
```

```python
df_vintrafamiliar.info()
```

** Normalizacin o estandarizacin de columnas.**

```python
#Normalizacin o estandarizacin de columnas.
df_vintrafamiliar.columns = df_vintrafamiliar.columns.str.lower()
df_vintrafamiliar
```

** Conversin de tipos de datos (por ejemplo: convertir fechas de string a formato datetime).**

```python
#  convertir fechas de string a formato datetime
df_vintrafamiliar['ao'] = pd.to_datetime(df_vintrafamiliar['ao'], errors='coerce')
print('tipo de datos actual de la columna ao:', df_vintrafamiliar['ao'].dtypes)

df_vintrafamiliar['estrato'] = pd.to_numeric(df_vintrafamiliar['estrato'], errors='coerce').astype('Int64')
print('tipo de datos actual de la columna  estrato :', df_vintrafamiliar['estrato'].dtypes)
```

```python
df_vintrafamiliar.info()
```

#COLUMNAS ELIMINADAS

```python
#Eliminar las columnas vacia
df_vintrafamiliar.drop(columns=['gestante'], inplace=True)
df_vintrafamiliar.drop(columns=['agresor_consumospa'], inplace=True)
df_vintrafamiliar.drop(columns=['victima_consumospa'], inplace=True)
df_vintrafamiliar.drop(columns=['lugocurrenciaabandono'], inplace=True)
df_vintrafamiliar.drop(columns=['lugocurrencianegligencia'], inplace=True)
df_vintrafamiliar.drop(columns=['lugocurrenciaeconomica'], inplace=True)
df_vintrafamiliar.drop(columns=['lugocurrenciaemocional'], inplace=True)
df_vintrafamiliar.drop(columns=['lugocurrenciafisica'], inplace=True)
df_vintrafamiliar.drop(columns=['lugocurrenciasexual'], inplace=True)
df_vintrafamiliar
```
