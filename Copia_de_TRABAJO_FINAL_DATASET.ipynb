```python
#Normalización o estandarización de columnas.
df_vintrafamiliar.columns = df_vintrafamiliar.columns.str.lower()
df_vintrafamiliar
```

**• Conversión de tipos de datos (por ejemplo: convertir fechas de string a formato datetime).**

```python
#  convertir fechas de string a formato datetime
df_vintrafamiliar['año'] = pd.to_datetime(df_vintrafamiliar['año'], errors='coerce')
print('tipo de datos actual de la columna año:', df_vintrafamiliar['año'].dtypes)

df_vintrafamiliar['estrato'] = pd.to_numeric(df_vintrafamiliar['estrato'], errors='coerce').astype('Int64')
print('tipo de datos actual de la columna  estrato :', df_vintrafamiliar['estrato'].dtypes)
```
